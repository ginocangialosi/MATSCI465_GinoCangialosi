## Exercise 1: Electron Wavelength Calculation

The relativistic wavelength of an electron is given by:
$$ \lambda = \frac{h}{\sqrt{2m_0eV(1 + \frac{eV}{2m_0c^2})}} $$

**Task:**
1.  Define constants: $h$ (Planck's constant), $m_0$ (electron mass), $e$ (elementary charge), $c$ (speed of light).
2.  Calculate the wavelength for an accelerating voltage $V = 200 \text{ kV}$.
3.  Print the result in picometers (pm).

```python
# Your code here
import scipy.constants as const

# Constants
h = const.h
m0 = const.m_e
e = const.e
c = const.c

V = 200e3 # 200 kV

# Calculate wavelength
w = (h) / (2 * m0 * e * V* (1+(e*V)/(2*m0*c**2) ))**(1/2)

a = w * 1e12
print(a)
```

## Exercise 2: Filename Parsing

You have a list of filenames from an experiment.
`filenames = ["image_001.tif", "image_002.tif", "calibration_01.dm4", "image_003.tif"]`

**Task:**
1.  Create a new list containing only the files that start with "image".
2.  Extract the number from each of these filenames (e.g., `1`, `2`, `3`).

```python
# Your code here
filenames = ["image_001.tif", "image_002.tif", "calibration_01.dm4", "image_003.tif"]

# Filter and extract
image_files = []

for name in filenames:
    if name.startswith("image"):
        number_part = name.split("_")[1].split(".")[0]
        image_files.append(int(number_part))

print(image_files)
```

## Exercise 3: NumPy Arrays and Statistics

In microscopy, we often deal with 2D arrays of pixel intensities.

**Task:**
1.  Create a 5x5 NumPy array of random integers between 0 and 255 (simulating a tiny image patch).
2.  Calculate the mean, minimum, and maximum intensity values.
3.  Normalize the array so that values range from 0.0 to 1.0.

```python
# Your code here
import numpy as np

# Create random array
arr = np.random.randint(0,256, size=(5,5))
print(arr)

# Calculate stats
mean = np.mean(arr)
min = np.amin(arr)
max = np.amax(arr)
print(mean, min, max)

# Normalize
arr_normalized = (arr - arr.min()) / (arr.max() - arr.min())
print(arr_normalized)
```

## Exercise 4: Simple Plotting

Visualizing data is crucial.

**Task:**
1.  Generate an array of x values from 0 to $4\pi$ using `np.linspace`.
2.  Calculate $y = \sin(x) + 0.1 \times \text{random\_noise}$.
3.  Plot $x$ vs $y$ using `matplotlib.pyplot`.
4.  Add labels for the x-axis ("Angle [rad]") and y-axis ("Amplitude").
5.  Add a title "Noisy Sine Wave".

```python
# Your code here
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0 , 4 * np.pi, 100)
noise = np.random.uniform(0, 4 * np.pi, x.shape)
y = np.sin(x) + 0.1 * noise

# Plot
plt.plot(x, y, label='y = sin(x) + 0.1 * random_noise', color='blue', marker='o')

# Labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Example Plot')
plt.legend()

# Show plot
plt.show()
```
